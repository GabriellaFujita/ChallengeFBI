```python
filtered_df = pandas.DataFrame()
filtered_df['id'] = df['entity_id']
filtered_df['nome'] = df['forename'] + ' ' + df['name']
filtered_df['nome'].fillna('Sem Nome', inplace=True)
filtered_df['url'] = df['link_id']
filtered_df.head()
```
