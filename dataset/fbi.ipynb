```python
import requests
import json
import pandas

i=0
df = pandas.DataFrame()
hasData = True

while(hasData):
    i += 1
    response = requests.get(f'https://api.fbi.gov/@wanted?pageSize=50&page={i}')
    data = json.loads(response.content)
    df = df.append(data['items'], ignore_index=True)
    if (df.shape[0] < 50*i - 1):
        hasData = False

df.head()
```

```python
filtered_df = pandas.DataFrame()
filtered_df['id'] = df['uid']
filtered_df['nome'] = df['title']
filtered_df['url'] = df['url']
filtered_df.head()
```

```python
filtered_df.iloc[1]
```

```python
filtered_df.to_json('fbi_data.json', orient='records', indent=4)
filtered_df.to_csv('fbi_data.csv', index=False)
```
